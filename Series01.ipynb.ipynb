# Series01 Python: Download and plot a ECG signal

```python
# we retrocode the code for the sake of this exercise
# to have a plot(ecg) we nedd to load the data with ecg=np.loadtxt("data/ECGu.txt"),we need to import numpy and matplotlib.pyplot as plt
#
import numpy as np
import matplotlib.pyplot as plt
## define the input file name and directory
input_file = "data/ECGu.txt"
## read the data 
ecg = np.loadtxt(input_file)
```

```python
print(ecg.shape)
#plot the ecg signal
plt.plot(ecg)
plt.title("ECG signal")
plt.xlabel("samples(rows)")
plt.ylabel("Amplitude")
plt.show()
```

```python
#we are going to rescale the ecg signal to a standard unit
#multiply the raw ecg data into 1024 µVolt
ECGraw=ecg
#conversion to second
fs=1000
n_sample=len(ecg)
T=1/fs
N=len(ecg)
start=0
stop=(N-1)*T
time_vec=np.linspace(start,stop,n_sample,endpoint=True)
#add legend outside the plot
#I give a label to each lead to be able to display it in the legend,I didnt manage to do it with a loop
plt.plot(time_vec, ecg[:, 0], label='ECG lead I')
plt.plot(time_vec, ecg[:, 1], label='ECG lead II')
plt.plot(time_vec, ecg[:, 2], label='ECG lead III')
plt.legend(loc='upper right', bbox_to_anchor=(1.15, 1))
#plot the ecg signal
plt.title("ECG signal")
plt.xlabel("time (s)")
plt.ylabel("ECG")
plt.grid()
plt.show()
```

```python
#rescale the ecg signal to a standard unit 
#multiply the raw ecg data into 1024 µVolt
ECGmv=ECGraw*1024e-6
#conversion to second
fs=1000
n_sample=ECGmv.shape[0]
T=1/fs
N=len(ECGmv)
start=0
stop=(N-1)*T
time_vec=np.linspace(start,stop,n_sample,endpoint=True)
#add legend outside the plot
#I give a label to each lead to be able to display it in the legend,I didnt manage to do it with a loop
plt.plot(time_vec, ecg[:, 0], label='ECG lead I')
plt.plot(time_vec, ecg[:, 1], label='ECG lead II')
plt.plot(time_vec, ecg[:, 2], label='ECG lead III')
plt.legend(loc='upper right', bbox_to_anchor=(1.15, 1))
#plot the ecg signal
plt.title("ECG signal")
plt.xlabel("time (s)")
plt.ylabel("ECG(mv)")
plt.grid()
plt.show()
```
