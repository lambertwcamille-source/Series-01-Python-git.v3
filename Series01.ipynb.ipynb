# Series01 Python: Download and plot a ECG signal

```python
# we retrocode the code for the sake of this exercise
# to have a plot(ecg) we nedd to load the data with ecg=np.loadtxt("data/ECGu.txt"),we need to import numpy and matplotlib.pyplot as plt
#
import numpy as np
import matplotlib.pyplot as plt
## define the input file name and directory
input_file = "data/ECGu.txt"
## read the data 
ecg = np.loadtxt(input_file)
```

```python
print(ecg.shape)
#plot the ecg signal
plt.plot(ecg)
plt.title("ECG signal")
plt.xlabel("samples(rows)")
plt.ylabel("Amplitude")
plt.show()
```

```python
#we are going to rescale the ecg signal to a standard unit
#multiply the raw ecg data into 1024 Volt
ECGraw=ecg
ECGmv=ECGraw*1024e-6

#conversion to second
fs=1000
n_sample=ECGmv.shape[0]
#time vector in seconds
t_start=0
n_sample=len(ecg)
t_ends=n_sample/fs
t_step=1/fs
time_vec=np.arange(t_start,t_ends,t_step)
#possible de le faire avec linspace(0, n_samples/fs, n_samples)
#add legend outside the plot
#I give a label to each lead to be able to display it in the legend,I didnt manage to do it with a loop
plt.plot(time_vec, ECGmv[:, 0], label='ECG lead I')
plt.plot(time_vec, ECGmv[:, 1], label='ECG lead II')
plt.plot(time_vec, ECGmv[:, 2], label='ECG lead III')
plt.legend(loc='upper right', bbox_to_anchor=(1.15, 1))
#plot the ecg signal
plt.title("ECG signal")
plt.xlabel("ECG (mv)")
plt.ylabel("Times(s)")
plt.grid()
plt.show()
```

```python
#we have to create the grid condition befeore the plot
import matplotlib.ticker as ticker
xmajor_locator = 0.2 
xminor_locator = 0.04
ymajor_locator = 0.5
yminor_locator = 0.1
#You have multiple leads (I, II, III) to display. You can create subplots for each lead to mimic the appearance of a real ECG monitor
fig, axes = plt.subplots(3, 1, figsize=(12, 6), sharex=True, sharey=True, )
axes[0].plot(time_vec, ECGmv[:, 0], color='blue',)
axes[0].set_title('I')
axes[0].grid(True,c='orange',which='both', axis='both')
axes[0].xaxis.set_major_locator(ticker.MultipleLocator(xmajor_locator))
axes[0].xaxis.set_minor_locator(ticker.MultipleLocator(xminor_locator))
axes[0].yaxis.set_major_locator(ticker.MultipleLocator(ymajor_locator))
axes[0].yaxis.set_minor_locator(ticker.MultipleLocator(yminor_locator))
axes[0].grid(which="major", color="orange", linestyle="-", linewidth=0.8)
axes[0].grid(which="minor", color="orange", linestyle="-", linewidth=0.5)
axes[1].plot(time_vec, ECGmv[:, 1], color='blue')
axes[1].set_title('II')
axes[1].grid(True,c='orange', linestyle='-')
axes[1].xaxis.set_major_locator(ticker.MultipleLocator(xmajor_locator))
axes[1].xaxis.set_minor_locator(ticker.MultipleLocator(xminor_locator))
axes[1].yaxis.set_major_locator(ticker.MultipleLocator(ymajor_locator))
axes[1].yaxis.set_minor_locator(ticker.MultipleLocator(yminor_locator))
axes[1].grid(which="major", color="orange", linestyle="-", linewidth=0.8)
axes[1].grid(which="minor", color="orange", linestyle="-", linewidth=0.5)
axes[2].plot(time_vec, ECGmv[:, 2], color='blue')
axes[2].set_title('III')
axes[2].grid(True,c='orange', linestyle='-')
axes[2].xaxis.set_major_locator(ticker.MultipleLocator(xmajor_locator))
axes[2].xaxis.set_minor_locator(ticker.MultipleLocator(xminor_locator))
axes[2].yaxis.set_major_locator(ticker.MultipleLocator(ymajor_locator))
axes[2].yaxis.set_minor_locator(ticker.MultipleLocator(yminor_locator))
axes[2].grid(which="major", color="orange", linestyle="-", linewidth=0.8)
axes[2].grid(which="minor", color="orange", linestyle="-", linewidth=0.5)
plt.xlabel("Time (s)")
plt.ylabel("ECG (mV)")


```
